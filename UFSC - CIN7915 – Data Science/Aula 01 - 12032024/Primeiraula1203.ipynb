##**Número 2**: Gerar dois dicionários Python com chaves string e valores inteiros. Transformá-los em séries Pandas, concatená-los, ordenar o dicionário/série resultante por valores crescentes e plotá-lo com Matplotlib na forma de Histograma

```python
dicionario_um = {
    'Zezinho' : '10',
    'Alice' : '5',
    'Mariazinha' : '6',
    'Pedro' : '9'
}

dicionario_um
```

```python
dicionario_dois = {
    'Gabriel' : '17',
    'Alicia' : '21',
    'Pedro' : '43'
}

dicionario_dois
```

```python
#Converter dicionários em séries
serie_um = pd.Series(dicionario_um)
serie_dois = pd.Series(dicionario_dois)
print(serie_um)
print(serie_dois)
```

```python
#Concatenar as séries
serie_dics_concatenados = pd.concat([serie_um, serie_dois])
serie_dics_concatenados
```

```python
#Ordenar por valores
serie_ordenada = serie_dics_concatenados.sort_values()
serie_ordenada
```

```python
plt.hist(serie_dics_concatenados)
plt.show()
```

```python
# Plotar histograma
plt.hist(serie_ordenada, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma dos valores ordenados')
plt.grid(True)
plt.show()
```
